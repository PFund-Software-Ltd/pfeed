# Historical crypto data with CryptoHFTData

Download exchange-specific trades through PFeed, then request the same source data as one-minute OHLCV bars. The free tier can be used without an API key.

```python
%pip install -q "pfeed[cryptohftdata]"
```

```python
import pfeed as pe

client = pe.CryptoHFTData()
client.list_exchanges()
```

```python
download_args = {
    "product": "BTC_USDT_PERP",
    "exchange": "binance_futures",
    "symbol": "BTCUSDT",
    "start_date": "2024-07-13",
    "end_date": "2024-07-13",
}
trades = client.market_feed.download(resolution="tick", **download_args).data.collect()
trades.head()
```

```python
bars = client.market_feed.download(resolution="1m", **download_args).data.collect()
bars.head()
```
